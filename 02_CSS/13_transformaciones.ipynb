# Transformaciones con CSS.

## Preliminares.

### La clase contenedor.

````python
%%html

<style>
    .contenedor {
       margin-top: 60px;
       margin-left: 60px;
       width: 200px;
       height:200px;
       background-color:#e9f0f2;
    }
</style>
````

### Muestra original.

````python
%%html
<style>
    #ejemplo {
        background-color:gold;
        width: 100px;
        height: 100px;
        color:red;
        }
</style>
<div class="contenedor">
    <div id="ejemplo">texto</div>
</div>
````

## Transformaciones con CSS.

Las hojas de estilo cuentan con la propeidad ```transform```, la cual permite transformar elementos mediante el uso de diversos métodos.

El origen en CSS es en la mitad del elemento. 

* *translate()*
    * *translateX()*
    * *translateY()*.
* *scale()*
    * *scaleX()*
    * *scaleY()*.
* *skew()*
    * *skewX()*
    * *skewY()*
* *matrix()*
* *rotate()*
* *rotateX()*
* *rotateY()*
* *rotateZ()*

La sintaxis es la siguiente:
```
<selector>{
transform: <método>(<atributos>);
}
```

### El método *translate()*.

````python
%%html
<style>
    #ejemplo_1 {
        background-color:gold;
        width: 100px;
        height: 100px;
        color:red;
        transform: translate(250px, 150px);    
        }
</style>
<div class="contenedor">
    <div id="ejemplo_1">texto</div>
</div>
````

### Los métodos *translateX()* y *translateX()*.

````python
%%html
<style>
    #ejemplo_2 {
        background-color:gold;
        width: 100px;
        height: 100px;
        color:red;
        transform: translateX(50px);
        transform: translateY(-50px);    
        }
</style>
<div class="contenedor">
    <div id="ejemplo_2">texto</div>
</div>
````

### El método *scale()*.

````python
%%html
<style>
    #ejemplo_3 {
        background-color:gold;
        color:red;
        width: 100px;
        height: 100px;
        transform: scale(0.5, 1.2); 
        }
</style>
<div class="contenedor">
    <div id="ejemplo_3">texto</div>
</div>
````

### Los métodos *scaleX()* y *scaleY()*.

````python
%%html
<style>
    #ejemplo_4 {
        background-color:gold;
        color:red;
        width: 100px;
        height: 100px;
        transform: scaleX(0.75);
        transform: scaleY(1,2); 
        }
</style>
<div class="contenedor">
    <div id="ejemplo_4">texto</div>
</div>
````

### El método *skew()*.

````python
%%html
<style>
    #ejemplo_5 {
        background-color:gold;
        color:red;
        width: 100px;
        height: 100px;
        transform: skew(15deg, 20deg);
        }
</style>
<div class="contenedor">
    <div id="ejemplo_5">texto</div>
</div>
````

### Los métodos *skewX()* y *skewY()* .

````python
%%html
<style>
    #ejemplo_6 {
        background-color:gold;
        color:red;
        width: 100px;
        height: 100px;
        transform: skewX(15deg);
        transform: skewY(20deg);
        }
</style>
<div class="contenedor">
    <div id="ejemplo_6">texto</div>
</div>
````

### El método *matrix()*.

```
transform: matrix (scaleX(), skewY(), skewX(), scaleY(), translateX(), translateY())
```

````python
%%html
<style>
    #ejemplo_7 {
        background-color:gold;
        color:red;
        width: 100px;
        height: 100px;
        transform: matrix(0.5, -1.07, 0.2, 0.2, 150, 0);
        }
</style>
<div class="contenedor">
    <div id="ejemplo_7">texto</div>
</div>
````

### El método *rotate()*.

````python
%%html
<style>
    #ejemplo_8 {
        background-color:gold;
        color:red;
        width: 100px;
        height: 100px;
        transform: rotate(45deg); 
        }
</style>
<div class="contenedor">
    <div id="ejemplo_8">texto</div>
</div>
````

### Métodos de rotaciónes en 3D.

También usan la regla *transform* para rotar en los ejes *x*, *y* y *z*.

* *rotateX()*
* *rotateY()*
* *rotateZ()*

````python
%%html
<style>
    #ejemplo_9 {
        background-color:gold;
        color:red;
        width: 100px;
        height: 100px;
        transform: rotateX(15deg);
        transform: rotateY(45deg);
        transform: rotateZ(15deg);
        }
    .contenedor {
        margin-left:60px;
    }
</style>
<div class="contenedor">
    <div id="ejemplo_9">texto</div>
</div>
````

## La propiedad ```transform-style```.

##  El atributo ```perspective```.

````python
%%html
<style>
.contenedor {
    perspective: 50px;
}
#ejemplo_10 {
    background-color:gold;
    color:red;
    width: 100px;
    height: 100px;
    transform-style: preserve-3d;
    transform: rotateX(25deg);
}

</style>

<div class="contenedor">
  <div id="ejemplo_10">texto</div>
</div>
````
